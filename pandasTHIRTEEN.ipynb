```python
import pandas as pd
customer=pd.read_csv(r'C:\Users\pc\Downloads\Restaurant - Customers.csv')
food = pd.read_csv(r'C:\Users\pc\Downloads\Restaurant - Foods.csv')
week1=pd.read_csv(r'C:\Users\pc\Downloads\Restaurant - Week 1 Sales.csv')
satisfaction=pd.read_csv(r'C:\Users\pc\Downloads\Restaurant - Week 1 Satisfaction.csv')
week2=pd.read_csv(r'C:\Users\pc\Downloads\Restaurant - Week 2 Sales.csv')
```

```python
week1.merge(week2,how='inner',on='Customer ID')
```

```python
week1.merge(week2,how='outer',on='Customer ID',suffixes=['week1','week2'])
```

```python
week1.merge(week2,how='outer',on='Customer ID',suffixes=['week1','week2'],
            indicator=True)
```

```python
#how much data from left table, how much from right table, and how much from both
data=week1.merge(week2,how='outer',on='Customer ID',suffixes=['week1','week2'],
                 indicator=True)
data['_merge'].value_counts()
```

```python
week1.merge(week2,how='left',on='Customer ID',suffixes=['week1','week2'])
```

```python
week1.merge(week2,how='right',on='Customer ID',suffixes=['week1','week2'])
```
