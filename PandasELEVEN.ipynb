```python
df = pd.DataFrame(columns=fortune.columns)

for sec, data in sector:
    hp = data.nlargest(1, "Profits")
    df = pd.concat([df, hp])  # Use pd.concat

df   
```

```python
df = pd.DataFrame(columns=fortune.columns)

for sec, data in sector:
    le = data.nsmallest(1, "Employees")
    df = pd.concat([df, le], ignore_index=True)  # Use concat instead of append
df
```
